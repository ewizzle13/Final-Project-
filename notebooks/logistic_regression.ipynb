```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, f1_score, recall_score
```

```python
data_dir = "../data/final_project_data_sp2026_L"

csv_files = [
    os.path.join(data_dir, "ids_0.csv"),
    os.path.join(data_dir, "ids_1.csv"),
    os.path.join(data_dir, "ids_2.csv"),
    os.path.join(data_dir, "ids_4_.csv"),
]

parquet_files = [
    os.path.join(data_dir, "ids_5.parquet"),
    os.path.join(data_dir, "ids_6.parquet"),
    os.path.join(data_dir, "ids_8.parquet"),
    os.path.join(data_dir, "ids_11.parquet"),
]

dfs = []

for f in csv_files:
    temp = pd.read_csv(f)
    temp.columns = temp.columns.str.strip()
    dfs.append(temp)

for f in parquet_files:
    temp = pd.read_parquet(f)
    temp.columns = temp.columns.str.strip()
    dfs.append(temp)

df = pd.concat(dfs, ignore_index=True, sort=False)

print(df["Label"].value_counts())
print(df["Label"].unique())
```

```python

def transform_data(df):
    df = df.drop_duplicates().copy()
    df["Label"] = df["Label"].str.strip()

    # remove Heartbleed
    df = df[df["Label"] != "Heartbleed"].copy()

    # binary labels
    df["Label"] = df["Label"].apply(
        lambda x: "benign" if x.lower() == "benign" else "attack"
    )
    
    # Drop known identifier columns
    cols_to_drop = ["Destination Port", "Timestamp", "Source IP", "Destination IP"]
    existing_drops = [col for col in cols_to_drop if col in df.columns]
    df = df.drop(columns=existing_drops)

    # Automatically drop columns that have only 1 unique value
    constant_cols = [col for col in df.columns if df[col].nunique() <= 1]
    df = df.drop(columns=constant_cols)

    print("\nBinary labels:")
    print(df["Label"].value_counts())

    X = df.drop(columns=["Label"])
    y = df["Label"]

    return X, y, df
def remove_outliers_iqr(df):
    df = df.copy()

    numeric_cols = df.select_dtypes(include=[np.number]).columns

    # calculate IQR correctly
    Q1 = df[numeric_cols].quantile(0.25)
    Q3 = df[numeric_cols].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    for col in numeric_cols:
        median = df[col].median()
        df[col] = np.where(
            (df[col] < lower_bound[col]) | (df[col] > upper_bound[col]),
            median,
            df[col]
        )

    return df
x, y, df_clean = transform_data(df)

print(x.shape)
print(y.shape)
```

```python
numeric_cols = x.select_dtypes(include=[np.number])

plt.figure(figsize=(20, 6))
numeric_cols.iloc[:, :10].boxplot()
plt.xticks(rotation=45)
plt.title("Boxplots of First 10 Numeric Features")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

print(y_test.value_counts())

# apply IQR only to training data to avoid data leakage
df_train = pd.concat([x_train, y_train], axis=1)
df_train = remove_outliers_iqr(df_train)

x_train = df_train.drop(columns=["Label"])
y_train = df_train["Label"]

# fill missing values
x_train = x_train.fillna(x_train.median())
x_test = x_test.fillna(x_train.median())

scaler = StandardScaler()

numeric_cols = x_train.select_dtypes(include=[np.number]).columns

x_train = x_train.copy()
x_test = x_test.copy()

x_train[numeric_cols] = scaler.fit_transform(x_train[numeric_cols])
x_test[numeric_cols] = scaler.transform(x_test[numeric_cols])

print(x_train.shape)
print(y_train.shape)

np.unique(y_train, return_counts=True)
```

```python
# 1. Temporarily convert labels to numbers for correlation
y_numeric = y_train.apply(lambda x: 1 if x == "attack" else 0)

# 2. Calculate correlation with the target
correlations = x_train.corrwith(y_numeric)

# 3. Sort by absolute correlation value
sorted_corr = correlations.abs().sort_values(ascending=False)

print("Top 10 features most correlated with an Attack:")
display(sorted_corr.head(10))
```

```python
lr = LogisticRegression(max_iter=2000, class_weight="balanced")

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, pos_label="attack")
recall = recall_score(y_test, y_pred, pos_label="attack")
precision = precision_score(y_test, y_pred, pos_label="attack")
report = classification_report(y_test, y_pred, target_names=["benign", "attack"])
cm = confusion_matrix(y_test, y_pred, labels=["benign", "attack"])
```

```python
print(f"Classification Report:\n{report}")
print(f"Accuracy: {acc:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"Recall: {recall:.4f}")
print(f"Precision: {precision:.4f}")
print("Confusion Matrix:")
print(cm)
```
