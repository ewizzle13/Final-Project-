```python
parquet_files = [
    "final_project_data_sp2026_L/ids_5.parquet",
    "final_project_data_sp2026_L/ids_6.parquet",
    "final_project_data_sp2026_L/ids_8.parquet",
    "final_project_data_sp2026_L/ids_11.parquet"
]

dfs = []
for f in parquet_files:
    temp = pd.read_parquet(f)
    temp.columns = temp.columns.str.strip()
    dfs.append(temp)

df = pd.concat(dfs, ignore_index=True, sort=False)
```
